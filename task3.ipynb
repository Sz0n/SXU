```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
moto = pd.read_csv('samochody1tys.csv')
```

```python
#Wykres funkcji linowej 
x = [1,2,3] # lista argumentów x
y = [4,6,5] # lista wartości

plt.plot(x,y,'r',label='nasza funkcja')
plt.title('Funkcja liniowa')
plt.xlabel('axis X')
plt.ylabel('axis Y')
plt.legend()
plt.grid(True)
```

```python
#Wykres funkcji kwadratowej
x = range(-10, 11)  # lista argumentów x
y = []  # lista wartości

for i in x:
    y.append(i**2)

plt.plot(x,y,'g-s',label='nasza funkcja')
plt.title('Funkcja kwadratowa')
plt.xlabel('axis X')
plt.ylabel('axis Y')
plt.legend()
plt.grid(True)
```

```python
#Wykres funkcji trzeciego rzędu
x = range(-10, 11)  # lista argumentów x
y = []  # lista wartości

for i in x:
    y.append(i**3)

plt.plot(x,y,'-v',label='nasza funkcja')
plt.title('Funkcja trzeciego rzędu')
plt.xlabel('axis X')
plt.ylabel('axis Y')
plt.legend()
plt.grid(True)
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
z = np.cos(x)   

plt.plot(x,y,'--',x,z,':')
plt.title('Funkcje trygonometryczne')
plt.xlabel('axis X')
plt.ylabel('axis Y')
plt.legend(['sin(x)','cos(x)'])
plt.grid(True)
```

```python
#Wykres słupkowy ilości samochodów danej marki 
brands = moto['marka'].value_counts()
brands.plot.bar(figsize=(15,5))
plt.title('Ilosc samochodów danej marki')
plt.xlabel('Marka')
plt.ylabel('Ilość samochodów')
plt.show()
```

```python
#Wykres kołowy ilości danych modeli samochodów marki Opel
brands = moto[moto.marka == 'Opel']
models = brands['model'].value_counts()
models.plot.pie(figsize=(30,10))
plt.title('Ilość danych modeli samochodów marki Opel')
plt.show()
```

```python
#Histogram roku produkcji samochodów(wykres ukazujący częstotliowść ukazywania się danej wartości)
production = moto['rok_produkcji']
histogram = production.hist()
histogram.set_title('Histogram roku produkcji samochodów')
histogram.set_xlabel('Rok produkcji')
histogram.set_ylabel('Ilość')
plt.show()
```
