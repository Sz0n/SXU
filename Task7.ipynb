```python
import numpy as np
import pandas as pd
```

```python
# funkcja aktywacyjna naszego neuronu
def activation_function(x):
    if x < 0:
        return 0
    else:
        return 1
```

```python
# Bramk AND
x1 = [0,1,0,1]
x2 = [0,0,1,1]
weights = [1,1,-1.5]

AND_function_value = [weights[2]+x1*weights[1]+weights[0]*x2 for x1,x2 in zip(x1,x2)]
AND_output = [activation_function(x) for x in AND_function_value]
AND_truth_table = pd.DataFrame({
    'X1':x1,
    'X2':x2,
    "X1's weight":weights[0],
    "X2's weight":weights[1],
    'bias': weights[2],
    '(-1.5+x1+x2)':AND_function_value,
    'output':AND_output

})
display(AND_truth_table)
```

```python
# Bramka OR
x1 = [0,1,0,1]
x2 = [0,0,1,1]
weights = [1,1,-0.5]

OR_function_value = [weights[2]+x1*weights[1]+weights[0]*x2 for x1,x2 in zip(x1,x2)]
OR_output = [activation_function(x) for x in OR_function_value]
OR_truth_table = pd.DataFrame({
    'X1':x1,
    'X2':x2,
    "X1's weight":weights[0],
    "X2's weight":weights[1],
    'bias': weights[2],
    '(-0.5+x1+x2)':OR_function_value,
    'output':OR_output

})
display(OR_truth_table)
```

```python
#Bramka NOT
x1 = [1,0]
weights = [1,-2]

NOT_function_value = [weights[0]+x1*weights[1] for x1 in x1]
NOT_output = [activation_function(x) for x in NOT_function_value]
NOT_truth_table = pd.DataFrame({
    'X1':x1,
    "X1's weight":weights[1],
    'bias': weights[0],
    '(1-2*x1)':NOT_function_value,
    'output':NOT_output

})
display(NOT_truth_table)
```

```python
#Bramka XNOR

x1 = [0,1,0,1]
x2 = [0,0,1,1]

weights_a1 = [-1,-1,0.5]
weights_a2 = [1,1,-1.5]
weights_a3 = [1,1,-0.5]

# a1 - (A’.B’)
# a1 = f( 0.5 – x1 – x2 )
a1_function_value = [weights_a1[2]+x1*weights_a1[1]+weights_a1[0]*x2 for x1,x2 in zip(x1,x2)]
a1_output = [activation_function(x) for x in a1_function_value]

# a2 - (A.B)
# a2 = f( -1.5 + x1 + x2 )
a2_function_value = [weights_a2[2]+x1*weights_a2[1]+weights_a2[0]*x2 for x1,x2 in zip(x1,x2)]
a2_output = [activation_function(x) for x in a2_function_value]

# a3 - (A’.B’)+(A.B)
# a3 = f( -0.5 + x1 + x2 )
a3_function_value = [weights_a3[2]+a1*weights_a3[1]+weights_a3[0]*a2 for a1,a2 in zip(a1_output,a2_output)]
a3_output = [activation_function(x) for x in a3_function_value]

first_layer_truth_table = pd.DataFrame({
    'X1':x1,
    'X2':x2,
    "a1: X1's weight":weights_a1[0],
    "a1: X2's weight":weights_a1[1],
    'a1: bias': weights_a1[2],
    'a1:(0.5-x1-x2)':a1_function_value,
    "a2: X1's weight":weights_a2[0],
    "a2: X2's weight":weights_a2[1],
    'a2: bias': weights_a2[2],
    'a2:(-1.5+x1+x2)':a2_function_value,
    'a1':a1_output,
    'a2':a2_output,
    })
XNOR_truth_table = pd.DataFrame({
    'a1':a1_output,
    'a2':a2_output,
    "a3: a1's weight":weights_a3[0],
    "a3: a2's weight":weights_a3[1],
    'a3: bias': weights_a3[2],
    'a3:(-0.5+x1+x2)':a3_function_value,
    'a3':a3_output
    })

display(first_layer_truth_table)

display(XNOR_truth_table)
```

```python
#Bramka XOR
#W artykule brak XOR-a więc znalazłem taki odpowiednik 
#Źródło: https://becominghuman.ai/neural-network-xor-application-and-fundamentals-6b1d539941ed

x1 = [0,1,0,1]
x2 = [0,0,1,1]

weights_a1 = [1,1,-0.5]
weights_a2 = [-1,-1,1.5]
weights_a3 = [1,1,-1.5]

# a1 - OR
# a1 = f( -0.5 + x1 + x2 )
a1_function_value = [weights_a1[2]+x1*weights_a1[1]+weights_a1[0]*x2 for x1,x2 in zip(x1,x2)]
a1_output = [activation_function(x) for x in a1_function_value]

# a2 - NAND
# a2 = f( 1.5 - x1 - x2 )
a2_function_value = [weights_a2[2]+x1*weights_a2[1]+weights_a2[0]*x2 for x1,x2 in zip(x1,x2)]
a2_output = [activation_function(x) for x in a2_function_value]

# a1 - XOR
# a1 = f( -1.5 + x1 + x2 )
a3_function_value = [weights_a3[2]+a1*weights_a3[1]+weights_a3[0]*a2 for a1,a2 in zip(a1_output,a2_output)]
a3_output = [activation_function(x) for x in a3_function_value]


first_layer_truth_table = pd.DataFrame({
    'X1':x1,
    'X2':x2,
    "a1: X1's weight":weights_a1[0],
    "a1: X2's weight":weights_a1[1],
    'a1: bias': weights_a1[2],
    'a1:(0.5-x1-x2)':a1_function_value,
    "a2: X1's weight":weights_a2[0],
    "a2: X2's weight":weights_a2[1],
    'a2: bias': weights_a2[2],
    'a2:(-1.5+x1+x2)':a2_function_value,
    'a1':a1_output,
    'a2':a2_output,
    })
XOR_truth_table = pd.DataFrame({
    'a1':a1_output,
    'a2':a2_output,
    "a3: a1's weight":weights_a3[0],
    "a3: a2's weight":weights_a3[1],
    'a3: bias': weights_a3[2],
    'a3:(-0.5+x1+x2)':a3_function_value,
    'a3':a3_output
    })

display(first_layer_truth_table)

display(XOR_truth_table)
```
